```python
import networkx as nx
import matplotlib.pyplot as plt
```

![image.png](https://github.com/sanfla/Visualisasi_Data/blob/main/Graph_Network/image.png?raw=true)

```python
kota = [
    'Bekasi', 'Depok', 'Bogor', 'Sukabumi', 'Cianjur',
    'Purwakarta', 'Subang', 'Bandung', 'Sumedang',
    'Majalengka', 'Cirebon', 'Garut', 'Tasikmalaya'
]

jalur = [
    ('Bekasi', 'Depok'),
    ('Depok', 'Bogor'),
    ('Bogor', 'Sukabumi'),
    ('Sukabumi', 'Cianjur'),
    ('Cianjur', 'Bandung'),
    ('Bandung', 'Purwakarta'),
    ('Purwakarta', 'Subang'),
    ('Subang', 'Sumedang'),
    ('Sumedang', 'Majalengka'),
    ('Majalengka', 'Cirebon'),
    ('Bandung', 'Sumedang'),
    ('Bandung', 'Garut'),
    ('Garut', 'Tasikmalaya'),
    ('Cianjur', 'Purwakarta'),
    ('Bogor', 'Cianjur'),
    ('Bandung', 'Subang')
]
```

```python
G = nx.Graph()
G.add_nodes_from(kota)
G.add_edges_from(jalur)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
```

```python
top_degree = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)
top_betweenness = sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)

print("Kota Terpenting Berdasarkan Degree Centrality:")
for kota, nilai in top_degree[:3]:
    print(f"- {kota}: {nilai:.3f}")

print("\nKota Terpenting Berdasarkan Betweenness Centrality:")
for kota, nilai in top_betweenness[:3]:
    print(f"- {kota}: {nilai:.3f}")
```

```python
node_sizes = [5000 * degree_centrality[node] for node in G.nodes()]

pos = nx.spring_layout(G, seed=42)

plt.figure(figsize=(12, 8))
nx.draw_networkx_edges(G, pos, edge_color='gray', width=1.5)
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='skyblue', alpha=0.9)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

plt.title("Visualisasi Jaringan Transportasi Kota Besar di Jawa Barat", fontsize=14)
plt.axis('off')
plt.show()
```
