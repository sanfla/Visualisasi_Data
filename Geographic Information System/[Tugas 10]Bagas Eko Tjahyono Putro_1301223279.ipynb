<div style="background-color:rgb(49, 98, 166); padding: 20px; border-radius: 10px;">
    <b>Plot Untuk 20 Provinsi dengan Kasus Kematian COVID-19 Indonesia Terbanyak</b>
</div>

```python
data_covid = pd.read_csv("Case by Provinces - IDN-COVID19 - Sheet2.csv")
data_covid.head()
```

```python
data_covid = data_covid.sort_values(by='Death_cases', ascending=False).head(20).reset_index(drop=True)
```

```python
data_covid
```

```python
data_covid.drop(
    columns=['Type', 
             'Features Type', 
             'ID-number', 
             'Province_code', 
             'Confirmed_cases', 
             'Recovered_cases', 
             'Features Geometry Type'
             ], inplace=True
)

data_covid[['Longitude', 'Latitude']] = data_covid['Features Geometry Coordinates'].str.split(',', expand=True).astype(float)

def create_point(row):
    return Point(row['Longitude'], row['Latitude'])

data_covid['Coordinates'] = data_covid.apply(create_point, axis=1)
```

```python
data_covid.head()
```

```python
geo_df = gpd.GeoDataFrame(data_covid, geometry='Coordinates')

fig, ax = plt.subplots(figsize=(14, 8))
geo_df.plot(ax=ax, marker='o', color='red', markersize=30)
plt.title('Kasus Kematian COVID-19 di Indonesia')
plt.show()
```
